```python
import matplotlib.pyplot as plt
```

```python
day= ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
no= [1,2,3,4,5,6,7]
```

alpha is used to change opacity

```python

colors= ["y", "c", "m" , "g", "c", "y", "r"]
sizes=[50,1000,10,20,34,50,200]
plt.scatter(day,no, c=colors, s=sizes, alpha=0.7)
plt.show()
```

```python
colors= ["y", "c", "m" , "g", "c", "y", "r"]
sizes=[50,1000,10,20,34,50,200]
plt.scatter(day,no, c=colors, s=sizes, marker="*")
t=plt.colorbar()
plt.show()
```

```python
day= ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
no= [1,2,3,4,5,6,7]
meetings=[1,5,7,2,6,3,4]

colors= ["y", "c", "m" , "g", "c", "y", "r"]
sizes=[50,1000,10,20,34,50,200]
plt.scatter(day,no, c=colors, s=sizes)
plt.scatter(day,meetings, c=colors, marker="*")
t=plt.colorbar()
plt.show()
```
