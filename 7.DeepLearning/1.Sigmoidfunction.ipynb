sigmoid(x) = 1 / (1 + exp(-x))

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
def der_sigmoid(x):
  return sigmoid(x) * (1- sigmoid(x))
```

```python
x_data = np.linspace(-10,10)
y_data = sigmoid(x_data)
dy_data = der_sigmoid(x_data)
```

```python
plt.plot(x_data, y_data, label='Sigmoid')
plt.plot(x_data, dy_data, label='Derivative of Sigmoid')
plt.title('Sigmoid Activation Function & Derivative')
plt.legend()
plt.grid()
plt.show()
```
